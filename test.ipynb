## Function definitions

```python
# This function can be used to select all columns that contain certain substrings
# Example usage: filter_index(df.columns,'nsusp') returns ['nsusp1', 'nsusp2', 'nsusp3', 'nsusp4', 'nsusp5', 'nsusp6', 'nsusp7']
def filter_index (indexes,keyword):
    filter_func = lambda s: keyword in s
    return [w for w in indexes if filter_func(w)]
```

## Exploring SAT Scores Data

```python
plt = df[filter_index(df.columns,'psat')].hist(figsize=(10, 8))
```

PSATs 5, 6, and 7 all have the same values. Collapsed during data cleaning for simplicity.

#### SAT Scores by School and Gender

```python
satsm = df.pivot_table('psatm7', rows='ssname7',cols='gender7')
# satsm = satsm.dropna().sort(['F','M']) # - drop schools with no scores
satsm = satsm.sort(['F','M'])
satsm.plot(figsize=(20,8),title='Math SAT Scores by School and gender',kind='bar',color='rb')
```

```python
satsv = df.pivot_table('psatv7', rows='ssname7',cols='gender7')
# satsv = satsv.dropna().sort(['F','M']) # - drop schools with no scores
satsv = satsv.sort(['F','M'])
satsv.plot(figsize=(20,8),title='Verbal SAT Scores by School and gender',kind='bar',color='rb')
```

#### SAT Scores Discrepancy by Gender and School

```python
satsm['SAT Math Diff'] = (satsm['M'] - satsm['F'])
satsm['SAT Verbal Diff'] = (satsv['M'] - satsv['F'])
satsm[['SAT Math Diff','SAT Verbal Diff']].plot(kind='bar',figsize=(15,6),title='Discrepancy on Math and Verbal SAT Scores by Gender')
```

```python
df['psattotal'] = (df['psatv7'] + df['psatm7']);
dfgender = df.groupby('gender7')
dfgender['psattotal'].describe()
```

```python
satsv['diff'] = (satsv['M'] - satsv['F'])
satsv.sort_index(by='diff')[::-1]['diff'].plot(kind='bar',figsize=(10,6),title='Discrepancy on Verbal SAT Scores by Gender',color='g')
```

```python
satsm['diff'] = (satsm['M'] - satsm['F'])
satsm.sort_index(by='diff')[::-1]['diff'].plot(kind='bar',figsize=(10,6),title='Discrepancy on Verbal SAT Scores by Gender')
```

* Several schools have no data: Longview, Rock Terrace, Transition, Stephen Knolls
* Boys > Girls for math, Boys < Girls for verbal; total scores almost identical
